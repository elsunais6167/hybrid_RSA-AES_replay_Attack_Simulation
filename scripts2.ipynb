```python
import random
import time
import threading

class Sensor:
    def __init__(self, sensor_id):
        self.sensor_id = sensor_id

    def generate_message(self):
        timestamp = time.time()
        data = random.randint(1, 100)
        return {"sensor_id": self.sensor_id, "timestamp": timestamp, "data": data}

class Server:
    def __init__(self, server_id):
        self.server_id = server_id

    def process_message(self, message):
        received_timestamp = time.time()
        print(f"Server {self.server_id} received: {message} - Received at {received_timestamp}")

def simulate_sensor(sensor, servers):
    for _ in range(100):
        target_server = random.choice(servers)
        message = sensor.generate_message()
        print(f"Sensor {sensor.sensor_id} sending: {message}")
        time.sleep(random.uniform(0.1, 0.5))  # Simulate variable transmission delay
        target_server.process_message(message)

def main():
    # Create 50 sensors and 2 servers
    sensors = [Sensor(sensor_id) for sensor_id in range(1, 51)]
    servers = [Server(server_id) for server_id in range(1, 3)]

    # Create threads for each sensor
    threads = []
    for sensor in sensors:
        thread = threading.Thread(target=simulate_sensor, args=(sensor, servers))
        threads.append(thread)
        thread.start()

    # Wait for all threads to finish
    for thread in threads:
        thread.join()

if __name__ == "__main__":
    main()
```
